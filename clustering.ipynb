```python
TheseColumns = ['job', 'marital', 'education', 'default', 'housing', 'loan','contact', 'month', 'day_of_week', 'poutcome', 'subscribed']
df[TheseColumns]
```

```python
for i in TheseColumns : 
    Dict = {i:j for i,j in zip(df[i].unique(),range(df[i].nunique()))}
    df[f'Enc_{i}']= df[i].map(Dict)
    df.drop([i],axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.cluster import KMeans
KMeansModel = KMeans(n_clusters=5,init='k-means++', random_state=33,algorithm= 'lloyd')
KMeansModel.fit(df) 
```

```python
KMeans(n_clusters=5, random_state=33)
pd.DataFrame(KMeansModel.cluster_centers_,columns = df.columns,index= ['ClusterA','ClusterB','ClusterC','ClusterD','ClusterE'])
```

```python
len(KMeansModel.labels_)
```

```python
pd.Series(KMeansModel.labels_).value_counts()
```

```python
from tqdm import tqdm
IntertiaDict = {}
for i in tqdm(range(2,16)) : 
    KMeansModel = KMeans(n_clusters=i,init='k-means++', random_state=33,algorithm= 'lloyd') 
    KMeansModel.fit(df)
    IntertiaDict[i] =   KMeansModel.inertia_
```

```python
IntertiaDict
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (7, 4))
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('SSD')
#----------------------------------------------------
#Defining Data
XGraph = IntertiaDict.keys()
yGraph = IntertiaDict.values()

#----------------------------------------------------
# Drawing plot graph No 1 . .  :   
plt.plot(XGraph,yGraph,linewidth=1,color= 'r', # can be : b ,g ,y ,c ,m ,k ,w 
         alpha=1,linestyle = 'solid') # can be  : dotted, dashed , dashdot




plt.show()
```
